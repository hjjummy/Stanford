ÏùºÎ∞òÏ†ÅÏúºÎ°ú T(y)=y Î°ú ÏÑ§Ï†ïÌïòÎäî Í≤ΩÏö∞Í∞Ä ÎßéÏúºÎ©∞, Ïù¥Îäî y ÏûêÏ≤¥ÎßåÏùÑ ÏùòÎØ∏

    Ï¶â, Îëê Î≤°ÌÑ∞ Í∞ÑÏùò ÎÇ¥Ï†Å Œ∑‚ä§T(y) Ïù¥ Í≥ÑÏÇ∞ Í∞ÄÎä•ÌïòÎ†§Î©¥, Îëê Î≤°ÌÑ∞Í∞Ä Ï†ïÌôïÌûà Í∞ôÏùÄ Ï∞®ÏõêÏùÑ Í∞ÄÏ†∏Ïïº ÏùòÎØ∏Í∞Ä ÏûàÎã§.
    
- **Îß§Í∞úÎ≥ÄÏàòÍ∞Ä ÎßéÏúºÎ©¥ Ï∂©Î∂Ñ ÌÜµÍ≥ÑÎüâÎèÑ ÎßéÏïÑÏïº ÌïúÎã§**
    
    ÎßåÏïΩ Î™®Îç∏Ïùò Îß§Í∞úÎ≥ÄÏàòÍ∞Ä ÎßéÎã§Î©¥, Îç∞Ïù¥ÌÑ∞Î°úÎ∂ÄÌÑ∞ Ï∂îÏ∂úÌï¥Ïïº Ìï† Ï∂©Î∂Ñ ÌÜµÍ≥ÑÎüâÎèÑ Í∑∏Ïóê ÏÉÅÏùëÌïòÏó¨ ÎßéÏïÑÏßÑÎã§. Î™®Îç∏Ïùò ÏûêÏú†ÎèÑÎ•º ÎäòÎ¶¨Î†§Î©¥ Ï∂©Î∂Ñ ÌÜµÍ≥ÑÎüâ Ïó≠Ïãú Ïù¥Ïóê ÎßûÏ∂∞ Ï¶ùÍ∞ÄÌï¥Ïïº ÌïúÎã§.
    

### **üîπ Í∏∞Î≥∏ Ï∏°Ï†ïÍ∞í h(y) Í≥º ÏûêÏó∞ Îß§Í∞úÎ≥ÄÏàòÏôÄÏùò Í¥ÄÍ≥Ñ**

    Í∑∏Îü¨ÎÇò h(y)Îäî Ï†ÑÏ≤¥ ÌôïÎ•† Î∂ÑÌè¨Î•º Íµ¨ÏÑ±ÌïòÎäî Îç∞ Ï§ëÏöîÌïú Ïó≠Ìï†ÏùÑ ÌïúÎã§.
    
- **ÏûêÏó∞ Îß§Í∞úÎ≥ÄÏàò Œ∑ ÏôÄ T(y) Í∞ÑÏùò Ïú†ÏùºÌïú ÏÉÅÌò∏ÏûëÏö©**
    
    Œ∑ÏôÄ T(y)Îäî ÏßÄÏàò Ìï®ÏàòÏùò ÏßÄÏàò Î∂ÄÎ∂ÑÏóêÏÑú ÏÑ†ÌòïÏ†ÅÏúºÎ°ú ÏÉÅÌò∏ÏûëÏö©ÌïòÎ©∞,
    
    Ïù¥Îäî ÏùºÎ∞òÌôî ÏÑ†Ìòï Î™®Îç∏(GLM) Í≥º Í∞ôÏùÄ Í∞úÎÖêÍ≥º ÏßÅÏ†ëÏ†ÅÏúºÎ°ú Ïó∞Í≤∞ÎêúÎã§

## üìå3. Î°úÍ∑∏ Î∂ÑÌï† Ìï®Ïàòa(Œ∑) Ïùò Ïó≠Ìï†

### üîπ Î°úÍ∑∏ Î∂ÑÌï† Ìï®ÏàòÏùò Ï†ïÏùò



Z(Œ∑)Îäî Ï†ïÍ∑úÌôî ÏÉÅÏàòÏù¥Î©∞,  $a(Œ∑)=logZ(Œ∑)$

Ïù¥ Ìï®ÏàòÎäî ÌôïÎ•†ÏùÑ Ï†ïÍ∑úÌôîÌïòÎäî Ïó≠Ìï†ÏùÑ ÌïòÎ©∞, ÌôïÎ•† Î∂ÑÌè¨Î•º Í≤∞Ï†ïÌïòÎäî ÌïµÏã¨ ÏöîÏÜå

- Î™®Îç∏Ïùò Îß§Í∞úÎ≥ÄÏàòÎ•º ÏµúÏ†ÅÌôîÌïòÎäî Îç∞ Ï§ëÏöîÌïú Ïó≠Ìï† ÏàòÌñâ
- Îã§ÏñëÌïú Î∂ÑÌè¨Î•º ÎèôÏùºÌïú Î∞©ÏãùÏúºÎ°ú ÌëúÌòÑ Í∞ÄÎä•
- Í∏∞Ïö∏Í∏∞ ÌïòÍ∞ïÎ≤ï(Gradient Descent) Îì±ÏùÑ Ï†ÅÏö©ÌïòÏó¨ ÌïôÏäµ Í∞ÄÎä•

### üîπ Î°úÍ∑∏ Î∂ÑÌï† Ìï®ÏàòÏùò ÌäπÏÑ±

**‚úÖ a(Œ∑)Îäî y Ïóê ÏùòÏ°¥ÌïòÏßÄ ÏïäÎäîÎã§. ‚Üí** Îî∞ÎùºÏÑú, ÌäπÏ†ï y Í∞íÏóê Îî∞Îùº Î≥ÄÌïòÏßÄ ÏïäÎäîÎã§.

**‚úÖ ÌôïÎ•† Î∂ÑÌè¨Ïùò Í∏∞ÎåÄÍ∞í Î∞è Î∂ÑÏÇ∞Í≥º Ïó∞Í≤∞Îê®**

- Î°úÍ∑∏ Î∂ÑÌï† Ìï®ÏàòÏùò 1Ï∞® ÎèÑÌï®Ïàò ‚Üí Í∏∞ÎåÄÍ∞í
- Î°úÍ∑∏ Î∂ÑÌï† Ìï®ÏàòÏùò 2Ï∞® ÎèÑÌï®Ïàò ‚Üí  Î∂ÑÏÇ∞



**‚úÖ Î™®Îì† Î∂àÏó∞ÏÜçÏ†ÅÏù∏ y Í∞íÏùò Ìï©ÏùÄ 1Ïù¥ ÎêòÏñ¥Ïïº Ìï®**

- Ï¶â, Îã§ÏùåÏùÑ ÎßåÏ°±Ìï¥Ïïº ÌïúÎã§.

$\sum_y P(y | \eta) = 1$

- Ïù¥Î•º ÌÜµÌï¥ ÌôïÎ•† Î∂ÑÌè¨Ïùò Ï†ïÍ∑úÌôîÍ∞Ä Î≥¥Ïû•ÎêúÎã§.

## üìå 4. Ïä§ÏπºÎùº Ìï®ÏàòÏôÄ Ï∞®Ïõê Ï†ïÌï©ÏÑ±

‚úÖ y,A(Œ∑),b(y) Îäî Î™®Îëê Ïä§ÏπºÎùº Ìï®ÏàòÏù¥Îã§.

‚úÖ ÏûêÏó∞ Îß§Í∞úÎ≥ÄÏàò Œ∑ ÏôÄ Ï∂©Î∂Ñ ÌÜµÍ≥ÑÎüâ T(y) Îäî ÎèôÏùºÌïú Ï∞®ÏõêÏùÑ Í∞ÄÏ†∏Ïïº ÌïúÎã§.

‚úÖ ÎßåÏïΩ Œ∑ Î•º p Î°ú Î≥ÄÌôòÌïòÎ©¥, Ïù¥Î•º A(p) Î°ú ÌëúÌòÑÌï† ÏàòÎèÑ ÏûàÎã§.

‚úÖ ÎßàÏ∞¨Í∞ÄÏßÄÎ°ú, y Ïùò Í∏∞Î≥∏ Ï∏°Ï†ïÍ∞íÎèÑ b(y) Î°ú Î≥ÄÍ≤ΩÌï† Ïàò ÏûàÎã§.

## üìå 5. ÏòàÏ†ú: ÎèôÏ†Ñ ÎçòÏßÄÍ∏∞ Î¨∏Ï†ú

ÎèôÏ†Ñ ÎçòÏßÄÍ∏∞ Î¨∏Ï†úÎ•º ÌÜµÌï¥ ÏßÄÏàò Í∞ÄÏ°± Î™®Îç∏Ïù¥ Ïñ¥ÎñªÍ≤å ÌôúÏö©ÎêòÎäîÏßÄ ÏÇ¥Ìé¥Î≥¥Ïûê.

### **‚úÖ Î¨∏Ï†ú ÏÑ§Ï†ï**

- ÎèôÏ†ÑÏùÑ ÎçòÏ°åÏùÑ Îïå ÏïûÎ©¥(y=1)Ïù¥ ÎÇòÏò¨ ÌôïÎ•†: p
- Îí∑Î©¥(y=0)Ïù¥ ÎÇòÏò¨ ÌôïÎ•†: 1‚àíp

Ïù¥Î•º **ÏßÄÏàò Í∞ÄÏ°± Î™®Îç∏**Î°ú ÌëúÌòÑÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÎã§:

$P(y)=exp(ylogp+(1‚àíy)log(1‚àíp))$

### **üß© Íµ¨ÏÑ± ÏöîÏÜå Î∂ÑÏÑù**

| Í∞úÎÖê | ÏàòÏãù |
| --- | --- |
| **ÏûêÏó∞ Îß§Í∞úÎ≥ÄÏàò(Natural Parameter)** | Œ∑=log(p/1‚àíp)‚Äã |
| **Ï∂©Î∂Ñ ÌÜµÍ≥ÑÎüâ(Sufficient Statistics)** | T(y)=y |
| **Î°úÍ∑∏ Î∂ÑÌï† Ìï®Ïàò(Log Partition Function)** | a(Œ∑)=‚àílog‚Å°(1‚àíp) |

MLEÎ•º ÏÇ¨Ïö©Ìïú ÎèôÏ†Ñ ÎçòÏßÄÍ∏∞ ÌôïÎ•† Ï∂îÏ†ï

```python
import numpy as np
import matplotlib.pyplot as plt

# Í∞ÄÏÉÅÏùò Îç∞Ïù¥ÌÑ∞: ÎèôÏ†Ñ 10Î≤à ÎçòÏ°åÏùÑ Îïå ÏïûÎ©¥(1)Ïù¥ ÎÇòÏò® ÌöüÏàò
data = np.array([1, 0, 1, 1, 0, 1, 1, 0, 1, 0])  # 1ÏùÄ ÏïûÎ©¥, 0ÏùÄ Îí∑Î©¥
n = len(data)

# Í∞ÄÎä•ÎèÑ Ìï®Ïàò L(p)
def likelihood(p, data):
    return np.prod(p**data * (1 - p)**(1 - data))

# Î°úÍ∑∏ Í∞ÄÎä•ÎèÑ Ìï®Ïàò ‚Ñì(p)
def log_likelihood(p, data):
    return np.sum(data * np.log(p) + (1 - data) * np.log(1 - p))

# Í∞ÄÎä•Ìïú p Í∞í Î≤îÏúÑ ÏÑ§Ï†ï
p_values = np.linspace(0.01, 0.99, 100)
likelihoods = [likelihood(p, data) for p in p_values]
log_likelihoods = [log_likelihood(p, data) for p in p_values]

# MLEÏóêÏÑú ÏµúÏ†ÅÏùò pÎäî Îç∞Ïù¥ÌÑ∞ÏóêÏÑú ÏïûÎ©¥Ïù¥ ÎÇòÏò® ÎπÑÏú®Í≥º Í∞ôÏùå
mle_p = np.mean(data)

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(p_values, likelihoods, label="Likelihood")
plt.axvline(mle_p, color='red', linestyle='--', label=f'MLE: p={mle_p:.2f}')
plt.xlabel("p")
plt.ylabel("Likelihood")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(p_values, log_likelihoods, label="Log Likelihood")
plt.axvline(mle_p, color='red', linestyle='--', label=f'MLE: p={mle_p:.2f}')
plt.xlabel("p")
plt.ylabel("Log Likelihood")
plt.legend()

plt.suptitle("MLE: Likelihood & Log-Likelihood")
plt.show()
```

```python
# Í∏∞Ïö∏Í∏∞ ÌïòÍ∞ïÎ≤ïÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ MLE ÏµúÏ†ÅÌôî
def gradient_descent(data, lr=0.1, epochs=100):
    p = 0.5  # Ï¥àÍ∏∞ ÌôïÎ•† ÏÑ§Ï†ï
    for _ in range(epochs):
        grad = np.sum(data) / len(data) - p  # Î°úÍ∑∏ Í∞ÄÎä•ÎèÑÏùò ÎØ∏Î∂Ñ
        p += lr * grad  # ÏóÖÎç∞Ïù¥Ìä∏
        p = np.clip(p, 0.01, 0.99)  # ÌôïÎ•†Ïù¥ 0~1 Î≤îÏúÑ Ïú†ÏßÄ
    return p

# ÏµúÏ†ÅÏùò p Í∞í Ï∞æÍ∏∞
mle_p_gd = gradient_descent(data)

print(f"MLE ÌôïÎ•† (Gradient Descent): {mle_p_gd:.4f}")
```

## üìå 6. ÏµúÎåÄ Í∞ÄÎä•ÎèÑ Ï∂îÏ†ï(MLE)Í≥º Ïò§Î™©ÏÑ±(Convexity)

### üîπ ÏµúÎåÄ Í∞ÄÎä•ÎèÑ Ï∂îÏ†ï(MLE)Ïùò Í∞úÎÖê

      MLE(Maximum Likelihood Estimation)ÏùÄ Î™®Îç∏Ïùò **Îß§Í∞úÎ≥ÄÏàòÎ•º ÏµúÏ†ÅÌôî**ÌïòÎäî Ï§ëÏöîÌïú Î∞©Î≤ï Ï§ë ÌïòÎÇò

     Î™©Ìëú: Ï£ºÏñ¥ÏßÑ Îç∞Ïù¥ÌÑ∞Î•º ÌÜµÌï¥ Î™®Îç∏Ïùò **Í∞ÄÎä•ÎèÑ(likelihood)Î•º ÏµúÎåÄÌôîÌïòÎäî Îß§Í∞úÎ≥ÄÏàò**Î•º Ï∞æÎäî Í≤É

### üîπ Í∞ÄÎä•ÎèÑ Ìï®Ïàò Ï†ïÏùò

- Ï£ºÏñ¥ÏßÑ Îç∞Ïù¥ÌÑ∞ **{ùë¶ùëñ}** (ùëñ=1, ‚Ä¶, ùëõ) Ïóê ÎåÄÌï¥ Í∞ÄÎä•ÎèÑ Ìï®Ïàò:  L(Œ∑) = ‚àè P(y·µ¢ | Œ∑)
- Î°úÍ∑∏ Í∞ÄÎä•ÎèÑ Ìï®ÏàòÎ°ú Î≥ÄÌôò:  ‚Ñì(Œ∑) = ‚àë (y·µ¢ Œ∑ ‚àí a(Œ∑))

- Î™©Ìëú: **‚Ñì(Œ∑)Î•º ÏµúÎåÄÌôîÌïòÎäî Œ∑ Ï∞æÍ∏∞**

### üîπ Ïò§Î™©ÏÑ±(Convexity) Í≤ÄÏ¶ù

‚úÖ **Í∞ÄÎä•ÎèÑ Ìï®ÏàòÍ∞Ä Ïò§Î™©(Convex)Ïù∏ÏßÄ ÌôïÏù∏ÌïòÎäî Í≥ºÏ†ï**

- 2Ï∞® ÎèÑÌï®ÏàòÎ•º Í≥ÑÏÇ∞ÌïòÎ©¥: d¬≤‚Ñì / dŒ∑¬≤ = - d¬≤a(Œ∑) / dŒ∑¬≤

- **Ïù¥ Í∞íÏù¥ Ìï≠ÏÉÅ ÏùåÏàòÎùºÎ©¥** ‚Üí **a(Œ∑)Îäî Î≥ºÎ°ù(Convex) Ìï®Ïàò**
- Îî∞ÎùºÏÑú, **ÏµúÏ†ÅÌôîÍ∞Ä Ïö©Ïù¥ÌïòÍ≥† ÏïàÏ†ïÏ†ÅÏù∏ Í≥ÑÏÇ∞ Í∞ÄÎä•**

## üìå 7. ÏßÄÏàò Í∞ÄÏ°± Î™®Îç∏Í≥º ÎπÑÏßÄÎèÑ ÌïôÏäµ

ÎπÑÏßÄÎèÑ ÌïôÏäµ **: ÎùºÎ≤®Ïù¥ ÏóÜÎäî Îç∞Ïù¥ÌÑ∞**Î•º ÏÇ¨Ïö©ÌïòÏó¨ **Îç∞Ïù¥ÌÑ∞Ïùò Íµ¨Ï°∞Î•º ÌïôÏäµ**ÌïòÎäî Î∞©Ïãù

 ÎåÄÌëúÏ†ÅÏù∏ Î∞©Î≤ï:

- **Íµ∞Ïßë Î∂ÑÏÑù(Clustering)**
- **Ï∞®Ïõê Ï∂ïÏÜå(Dimensionality Reduction)**
- **Î∞ÄÎèÑ Ï∂îÏ†ï(Density Estimation)**

**ÏßÄÏàò Í∞ÄÏ°± Î™®Îç∏ÏùÄ ÎπÑÏßÄÎèÑ ÌïôÏäµÏóêÏÑúÎèÑ Ï§ëÏöîÌïú Ïó≠Ìï†ÏùÑ Ìï®**

1Ô∏è‚É£ **Ïû†Ïû¨ Î≥ÄÏàò Î™®Îç∏(Latent Variable Models)**

- Í¥ÄÏ∏°ÎêòÏßÄ ÏïäÏùÄ Ïû†Ïû¨ Î≥ÄÏàò(latent variable) Í∞Ä Ìè¨Ìï®Îêú Îç∞Ïù¥ÌÑ∞ Î™®Îç∏ÎßÅ
- **Ïòà: Í∞ÄÏö∞ÏãúÏïà ÌòºÌï© Î™®Îç∏(GMM)**

2Ô∏è‚É£ **EM ÏïåÍ≥†Î¶¨Ï¶ò(Expectation-Maximization Algorithm)**

- ÏµúÎåÄ Í∞ÄÎä•ÎèÑ Ï∂îÏ†ï(MLE)ÏùÑ ÏàòÌñâÌïòÎäî Ìö®Ïú®Ï†ÅÏù∏ ÏïåÍ≥†Î¶¨Ï¶ò
- **Îç∞Ïù¥ÌÑ∞Ïùò Ïû†Ïû¨ Íµ¨Ï°∞Î•º Ï∂îÏ†ïÌïòÎäî Îç∞ Ïú†Ïö©**

EM ÏïåÍ≥†Î¶¨Ï¶òÏùÑ ÏÇ¨Ïö©Ìïú Í∞ÄÏö∞ÏãúÏïà ÌòºÌï© Î™®Îç∏(GMM)

```python
from sklearn.mixture import GaussianMixture
import seaborn as sns

# Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
np.random.seed(42)
data1 = np.random.normal(loc=2, scale=1, size=100)
data2 = np.random.normal(loc=7, scale=1.5, size=100)
data = np.concatenate([data1, data2]).reshape(-1, 1)

# GMM Î™®Îç∏ ÌïôÏäµ
gmm = GaussianMixture(n_components=2, random_state=42)
gmm.fit(data)

# ÏòàÏ∏°Îêú Íµ∞Ïßë Ï†ïÎ≥¥
labels = gmm.predict(data)

# Í≤∞Í≥º ÏãúÍ∞ÅÌôî
sns.histplot(data.ravel(), bins=30, kde=True, color='gray', alpha=0.5)
sns.scatterplot(x=data.ravel(), y=[0]*len(data), hue=labels, palette="coolwarm", s=50)
plt.title("Gaussian Mixture Model (GMM) Clustering")
plt.show()
```

## üìå 8. ÏùºÎ∞òÌôî ÏÑ†Ìòï Î™®Îç∏(Generalized Linear Models, GLM)

### üîπ GLMÏùò Í∞úÎÖê

‚úÖ GLMÏùÄ **ÏÑ§Í≥Ñ ÏÑ†ÌÉù(Design Choices) ÎòêÎäî Í∞ÄÏ†ï(Assumptions)Ïóê Í∏∞Î∞ò**

‚úÖ Îç∞Ïù¥ÌÑ∞ ÌäπÏÑ±Ïóê Îî∞Îùº **Ïò§Î•ò Î™®Îìú(Error Mode)Î•º Í≤∞Ï†ï**

### üîπ Îç∞Ïù¥ÌÑ∞ Ïú†ÌòïÎ≥Ñ GLM ÏÑ†ÌÉù

| Îç∞Ïù¥ÌÑ∞ Ïú†Ìòï | Ï†ÅÌï©Ìïú ÌôïÎ•† Î∂ÑÌè¨ |
| --- | --- |
| **Ïù¥ÏßÑ Îç∞Ïù¥ÌÑ∞(Binary Data)** | **Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨(Bernoulli Distribution)** |

## üìå 9. Î°úÏßÄÏä§Ìã± ÌöåÍ∑Ä(Logistic Regression)ÏôÄ ÌôïÎ•†Ï†Å ÏòàÏ∏°

### üîπ Î°úÏßÄÏä§Ìã± ÌöåÍ∑ÄÏùò Î∞úÏ†Ñ

‚úÖ Í≥ºÍ±∞ÏóêÎäî **Ïù¥ÏßÑ Í≤∞Í≥º(Yes/No)** Î•º Î∞òÌôò

‚úÖ ÌòÑÏû¨Îäî **ÌôïÎ•† Î∂ÑÌè¨Î•º Î∞òÌôò**ÌïòÏó¨ **ÌäπÏ†ï ÌÅ¥ÎûòÏä§Ïùº ÌôïÎ•†ÏùÑ Ï∂îÏ†ï**

‚úÖ ÌôïÎ•†Ï†Å ÏòàÏ∏° Î∞©ÏãùÏù¥ Î®∏Ïã†Îü¨ÎãùÏóêÏÑú Ï§ëÏöîÌïú ÏöîÏÜå

### üîπ ÏûêÏó∞ Îß§Í∞úÎ≥ÄÏàò Í≥µÍ∞Ñ(Natural Parameter Space)

‚úÖ Îç∞Ïù¥ÌÑ∞Î•º **ÏûêÏó∞ Îß§Í∞úÎ≥ÄÏàò Í≥µÍ∞Ñ**ÏúºÎ°ú Î≥ÄÌôòÌïòÎ©¥ **ÏµúÏ†ÅÌôîÍ∞Ä Îã®ÏàúÌôîÎê®**

‚úÖ Ïù¥ Í≥µÍ∞ÑÏóêÏÑú **Í∏∞Ïö∏Í∏∞ ÌïòÍ∞ïÎ≤ï(Gradient Descent)** ÏùÑ ÏàòÌñâ

Î°úÏßÄÏä§Ìã± ÌöåÍ∑ÄÎ•º ÏÇ¨Ïö©Ìïú Î∂ÑÎ•ò ÏòàÏ†ú

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (Ïò¨Î∞îÎ•∏ ÏÑ§Ï†ï)
X, y = make_classification(n_samples=100, n_features=1, n_informative=1, 
                           n_redundant=0, n_clusters_per_class=1, random_state=42)

# Î°úÏßÄÏä§Ìã± ÌöåÍ∑Ä Î™®Îç∏ ÌõàÎ†®
model = LogisticRegression()
model.fit(X, y)

# ÏòàÏ∏° ÌôïÎ•† ÌôïÏù∏
X_test = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_prob = model.predict_proba(X_test)[:, 1]

# Í≤∞Í≥º ÏãúÍ∞ÅÌôî
plt.scatter(X, y, label="Data", edgecolors='k')
plt.plot(X_test, y_prob, label="Logistic Regression", color="red")
plt.xlabel("Feature")
plt.ylabel("Probability")
plt.legend()
plt.title("Logistic Regression Decision Boundary")
plt.show()
```

# 지수 가족 모델 Exponential Family model

## 📌  1. 정의

### 🔹 지수 가족 모델의 핵심 아이디어

> ‘’if has special form ⇒ some  inference/ learning for free. ”
특정한 형태를 가지는 확률 분포라면, 학습 및 추론 과정에서 이점을 얻을 수 있음.
> 

### 🔹 수식 **형태**

$P(y \mid \eta) = h(y) \exp(\eta^\top T(y) - A(\eta))$

- y : 관측된 데이터
- η (Eta) : 자연 매개변수 (Natural Parameter)
- T(y) : 충분 통계량 (Sufficient Statistic)
- a(η) : 로그 분할 함수 (Log Partition Function), 확률을 정규화하는 역할
- h(y): 기본 측정값 (Base Measure), 자연 매개변수 η 와 독립적임

### **🔹 왜 지수 가족 모델이 중요한가?**

✅ **더 광범위한 오류를 처리 가능** → 다양한 데이터 분포와 특성을 다룰 수 있음

✅ **비지도 학습에 활용 가능** → 라벨이 없는 데이터에서도 학습 가능

✅ **머신러닝에서 반복적으로 사용되는 "서브루틴" 역할** → 복잡한 문제를 단순화

## 📌2. 자연 매개변수η와 관계

### 🔹 충분 통계량 T(y)

지수 가족 모델에서는 충분 통계량 T(y) 가 데이터를 요약하는 역할을 한다.

일반적으로 T(y)=y 로 설정하는 경우가 많으며, 이는 y 자체만을 의미

- **자연 매개변수와 동일한 차원**
    
    T(y)와 η는 동일한 차원을 가져야 한다.
    
    즉, 두 벡터 간의 내적 η⊤T(y) 이 계산 가능하려면, 두 벡터가 정확히 같은 차원을 가져야 의미가 있다.
    
- **매개변수가 많으면 충분 통계량도 많아야 한다**
    
    만약 모델의 매개변수가 많다면, 데이터로부터 추출해야 할 충분 통계량도 그에 상응하여 많아진다. 모델의 자유도를 늘리려면 충분 통계량 역시 이에 맞춰 증가해야 한다.
    

### **🔹 기본 측정값 h(y) 과 자연 매개변수와의 관계**

- **기본 측정값(Base Measure) h(y)**
    
    h(y)는 자연 매개변수 η에 직접적으로 영향을 받지 않는다.
    
    그러나 h(y)는 전체 확률 분포를 구성하는 데 중요한 역할을 한다.
    
- **자연 매개변수 η 와 T(y) 간의 유일한 상호작용**
    
    η와 T(y)는 지수 함수의 지수 부분에서 선형적으로 상호작용하며,
    
    이는 일반화 선형 모델(GLM) 과 같은 개념과 직접적으로 연결된다

## 📌3. 로그 분할 함수a(η) 의 역할

### 🔹 로그 분할 함수의 정의



Z(η)는 정규화 상수이며,  $a(η)=logZ(η)$

이 함수는 확률을 정규화하는 역할을 하며, 확률 분포를 결정하는 핵심 요소

- 모델의 매개변수를 최적화하는 데 중요한 역할 수행
- 다양한 분포를 동일한 방식으로 표현 가능
- 기울기 하강법(Gradient Descent) 등을 적용하여 학습 가능

### 🔹 로그 분할 함수의 특성

**✅ a(η)는 y 에 의존하지 않는다. →** 따라서, 특정 y 값에 따라 변하지 않는다.

**✅ 확률 분포의 기대값 및 분산과 연결됨**

- 로그 분할 함수의 1차 도함수 → 기대값
- 로그 분할 함수의 2차 도함수 →  분산



**✅ 모든 불연속적인 y 값의 합은 1이 되어야 함**

- 즉, 다음을 만족해야 한다.

$\sum_y P(y | \eta) = 1$

- 이를 통해 확률 분포의 정규화가 보장된다.

## 📌 4. 스칼라 함수와 차원 정합성

✅ y,A(η),b(y) 는 모두 스칼라 함수이다.

✅ 자연 매개변수 η 와 충분 통계량 T(y) 는 동일한 차원을 가져야 한다.

✅ 만약 η 를 p 로 변환하면, 이를 A(p) 로 표현할 수도 있다.

✅ 마찬가지로, y 의 기본 측정값도 b(y) 로 변경할 수 있다.

## 📌 5. 예제: 동전 던지기 문제

동전 던지기 문제를 통해 지수 가족 모델이 어떻게 활용되는지 살펴보자.

### **✅ 문제 설정**

- 동전을 던졌을 때 앞면(y=1)이 나올 확률: p
- 뒷면(y=0)이 나올 확률: 1−p

이를 **지수 가족 모델**로 표현하면 다음과 같다:

$P(y)=exp(ylogp+(1−y)log(1−p))$

### **🧩 구성 요소 분석**

| 개념 | 수식 |
| --- | --- |
| **자연 매개변수(Natural Parameter)** | η=log(p/1−p)​ |
| **충분 통계량(Sufficient Statistics)** | T(y)=y |
| **로그 분할 함수(Log Partition Function)** | a(η)=−log⁡(1−p) |

MLE를 사용한 동전 던지기 확률 추정

```python
import numpy as np
import matplotlib.pyplot as plt

# 가상의 데이터: 동전 10번 던졌을 때 앞면(1)이 나온 횟수
data = np.array([1, 0, 1, 1, 0, 1, 1, 0, 1, 0])  # 1은 앞면, 0은 뒷면
n = len(data)

# 가능도 함수 L(p)
def likelihood(p, data):
    return np.prod(p**data * (1 - p)**(1 - data))

# 로그 가능도 함수 ℓ(p)
def log_likelihood(p, data):
    return np.sum(data * np.log(p) + (1 - data) * np.log(1 - p))

# 가능한 p 값 범위 설정
p_values = np.linspace(0.01, 0.99, 100)
likelihoods = [likelihood(p, data) for p in p_values]
log_likelihoods = [log_likelihood(p, data) for p in p_values]

# MLE에서 최적의 p는 데이터에서 앞면이 나온 비율과 같음
mle_p = np.mean(data)

# 그래프 그리기
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(p_values, likelihoods, label="Likelihood")
plt.axvline(mle_p, color='red', linestyle='--', label=f'MLE: p={mle_p:.2f}')
plt.xlabel("p")
plt.ylabel("Likelihood")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(p_values, log_likelihoods, label="Log Likelihood")
plt.axvline(mle_p, color='red', linestyle='--', label=f'MLE: p={mle_p:.2f}')
plt.xlabel("p")
plt.ylabel("Log Likelihood")
plt.legend()

plt.suptitle("MLE: Likelihood & Log-Likelihood")
plt.show()
```

```python
# 기울기 하강법을 사용하여 MLE 최적화
def gradient_descent(data, lr=0.1, epochs=100):
    p = 0.5  # 초기 확률 설정
    for _ in range(epochs):
        grad = np.sum(data) / len(data) - p  # 로그 가능도의 미분
        p += lr * grad  # 업데이트
        p = np.clip(p, 0.01, 0.99)  # 확률이 0~1 범위 유지
    return p

# 최적의 p 값 찾기
mle_p_gd = gradient_descent(data)

print(f"MLE 확률 (Gradient Descent): {mle_p_gd:.4f}")
```

## 📌 6. 최대 가능도 추정(MLE)과 오목성(Convexity)

### 🔹 최대 가능도 추정(MLE)의 개념

      MLE(Maximum Likelihood Estimation)은 모델의 **매개변수를 최적화**하는 중요한 방법 중 하나

     목표: 주어진 데이터를 통해 모델의 **가능도(likelihood)를 최대화하는 매개변수**를 찾는 것

### 🔹 가능도 함수 정의

- 주어진 데이터 **{𝑦𝑖}** (𝑖=1, …, 𝑛) 에 대해 가능도 함수:  L(η) = ∏ P(yᵢ | η)
- 로그 가능도 함수로 변환:  ℓ(η) = ∑ (yᵢ η − a(η))

- 목표: **ℓ(η)를 최대화하는 η 찾기**

### 🔹 오목성(Convexity) 검증

✅ **가능도 함수가 오목(Convex)인지 확인하는 과정**

- 2차 도함수를 계산하면: d²ℓ / dη² = - d²a(η) / dη²

- **이 값이 항상 음수라면** → **a(η)는 볼록(Convex) 함수**
- 따라서, **최적화가 용이하고 안정적인 계산 가능**

## 📌 7. 지수 가족 모델과 비지도 학습

비지도 학습 **: 라벨이 없는 데이터**를 사용하여 **데이터의 구조를 학습**하는 방식

 대표적인 방법:

- **군집 분석(Clustering)**
- **차원 축소(Dimensionality Reduction)**
- **밀도 추정(Density Estimation)**

**지수 가족 모델은 비지도 학습에서도 중요한 역할을 함**

1️⃣ **잠재 변수 모델(Latent Variable Models)**

- 관측되지 않은 잠재 변수(latent variable) 가 포함된 데이터 모델링
- **예: 가우시안 혼합 모델(GMM)**

2️⃣ **EM 알고리즘(Expectation-Maximization Algorithm)**

- 최대 가능도 추정(MLE)을 수행하는 효율적인 알고리즘
- **데이터의 잠재 구조를 추정하는 데 유용**

EM 알고리즘을 사용한 가우시안 혼합 모델(GMM)

```python
from sklearn.mixture import GaussianMixture
import seaborn as sns

# 데이터 생성
np.random.seed(42)
data1 = np.random.normal(loc=2, scale=1, size=100)
data2 = np.random.normal(loc=7, scale=1.5, size=100)
data = np.concatenate([data1, data2]).reshape(-1, 1)

# GMM 모델 학습
gmm = GaussianMixture(n_components=2, random_state=42)
gmm.fit(data)

# 예측된 군집 정보
labels = gmm.predict(data)

# 결과 시각화
sns.histplot(data.ravel(), bins=30, kde=True, color='gray', alpha=0.5)
sns.scatterplot(x=data.ravel(), y=[0]*len(data), hue=labels, palette="coolwarm", s=50)
plt.title("Gaussian Mixture Model (GMM) Clustering")
plt.show()
```

3️⃣ **약한 데이터 소스 활용**

- 최근 소량의 라벨 데이터를 활용하여 더 강력한 모델을 학습
- 비지도 학습과 지도 학습의 결합 형태로 발전

## 📌 9. 로지스틱 회귀(Logistic Regression)와 확률적 예측

### 🔹 로지스틱 회귀의 발전

✅ 과거에는 **이진 결과(Yes/No)** 를 반환

✅ 현재는 **확률 분포를 반환**하여 **특정 클래스일 확률을 추정**

✅ 확률적 예측 방식이 머신러닝에서 중요한 요소

### 🔹 자연 매개변수 공간(Natural Parameter Space)

✅ 데이터를 **자연 매개변수 공간**으로 변환하면 **최적화가 단순화됨**

✅ 이 공간에서 **기울기 하강법(Gradient Descent)** 을 수행

로지스틱 회귀를 사용한 분류 예제

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성 (올바른 설정)
X, y = make_classification(n_samples=100, n_features=1, n_informative=1, 
                           n_redundant=0, n_clusters_per_class=1, random_state=42)

# 로지스틱 회귀 모델 훈련
model = LogisticRegression()
model.fit(X, y)

# 예측 확률 확인
X_test = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_prob = model.predict_proba(X_test)[:, 1]

# 결과 시각화
plt.scatter(X, y, label="Data", edgecolors='k')
plt.plot(X_test, y_prob, label="Logistic Regression", color="red")
plt.xlabel("Feature")
plt.ylabel("Probability")
plt.legend()
plt.title("Logistic Regression Decision Boundary")
plt.show()
```
